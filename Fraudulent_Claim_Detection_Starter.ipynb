#### **4.4.1** Target likelihood analysis for categorical variables. <font color="red">[5 Marks]</font>
Investigate the relationships between categorical features and the target variable by analysing the target event likelihood (for the `'Y'` event) for each level of every relevant categorical feature. Through this analysis, identify categorical features that do not contribute much in explaining the variation in the target variable.

```python
# Write a function to calculate and analyse the target variable likelihood for categorical features
def categorical_target_likelihood_df(df, target_col='fraud_reported', showflag=True, event_value='Y',threshold=0.001):
   
        # Select categorical columns (excluding target)
        cat_cols = df.select_dtypes(include=['object', 'category']).columns.drop(target_col)

        event_rate_variances = []
        result_df = pd.DataFrame()
        for col in cat_cols:
            # Calculate event rate for the event_value
            event_rate = df.groupby(col)[target_col].value_counts(normalize=True).unstack().fillna(0).get(event_value, pd.Series())

            if event_rate.empty:
                continue  # skip if no 'event_value' in this column

            # Store results for each category
            for cat, likelihood in event_rate.items():
                event_rate_variances.append({
                    "Features": col,
                    "Categories": cat,
                    "LikelyHood": likelihood,
                    "Event_Rate_Variance": event_rate.var()
                })
            if showflag:
                # Plot
                plt.figure(figsize=(8, 4))
                sns.barplot(x=event_rate.index, y=event_rate.values, palette='plasma')
                plt.title(f"Event Rate ('{event_value}') by {col}")
                plt.ylabel(f"Proportion of '{event_value}'")
                plt.xlabel(col)
                plt.xticks(rotation=55)
                plt.tight_layout()
                plt.show()
        

         # Sort and identify low-variance features
        result_df = pd.DataFrame(event_rate_variances, columns=["Features","Categories","LikelyHood","Event_Rate_Variance"])
        low_info_features = result_df[result_df["Event_Rate_Variance"] < threshold].sort_values('Event_Rate_Variance',ascending=False)
        print(f"Low-informative categorical lenght and features:{len(low_info_features)}", low_info_features["Features"].unique().tolist())
        
        return low_info_features
```

### **6.1 Perform resampling** <font color="red">[3 Marks]</font>

Handle class imbalance in the training data by applying resampling technique.

Use the **RandomOverSampler** technique to balance the data and handle class imbalance. This method increases the number of samples in the minority class by randomly duplicating them, creating synthetic data points with similar characteristics. This helps prevent the model from being biased toward the majority class and improves its ability to predict the minority class more accurately.

**Note:** You can try other resampling techniques to handle class imbalance

```python
# Import RandomOverSampler from imblearn library
from imblearn.over_sampling import RandomOverSampler

# Perform resampling on training data
ros = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = ros.fit_resample(X_train, y_train)

#Check new class distribution
from collections import Counter
print("Original training target distribution:", Counter(y_train))
print("Resampled training target distribution:", Counter(y_train_resampled))
```

```python
# Create new features based on your understanding for both training
def create_new_features(df):
    new_features=pd.DataFrame({})

    # Policy csl levels segmentation
    if 'policy_csl' in df.columns:
        new_features['policy_csl_levels']=pd.cut(
                                            df['policy_csl'],
                                            bins=[0,500,1000,df['policy_csl'].max()],
                                            labels=['Low','Medium','High'])


    # customer_tenure_level: Bucket or bin
    if 'months_as_customer' in df.columns:
        new_features['customer_tenure_level'] = pd.cut(
                                                    df['months_as_customer'],
                                                    bins=[0, 12, 36, 60, df['months_as_customer'].max()],  # ensures monotonicity
                                                    labels=['New', '1-3yrs', '3-5yrs', 'Long-term']
                                                )


    # Extract date parts
    if 'incident_date' in df.columns and 'incident_datetime' in df.columns:
        new_features['incident_year'] = df['incident_datetime'].dt.year
        new_features['incident_month'] = df['incident_datetime'].dt.month
        new_features['incident_dayofweek'] = df['incident_date'].dt.dayofweek
        new_features['incident_weekday'] = new_features['incident_dayofweek'].apply(lambda x: 0 if x >= 5 else 1)
        new_features['incident_hour'] = df['incident_datetime'].dt.hour


    #total_claim_ratio 
    if 'total_claim_amount' in df.columns and 'injury_claim' in df.columns and 'property_claim' in df.columns and 'vehicle_claim' in df.columns:
        new_features['total_claim_ratio'] = (
            df['total_claim_amount'] / 
            (df['injury_claim'] + df['property_claim'] + df['vehicle_claim'] + 1e-5)  # avoid division by zero
        )

    # umbrella_limit_flag
    if 'umbrella_limit' in df.columns:
        new_features['umbrella_limit_flag'] = df['umbrella_limit'].apply(lambda x: 1 if x > 0 else 0)



    # combine age and policy deductible to see if higher deductible impacts older customers
    if 'age' in df.columns and 'policy_deductable' in df.columns:
        new_features['age_deductible_interaction'] = df['age'] * df['policy_deductable']

    # total_claim_amount / months_as_customer ratio
    if 'total_claim_amount' in df.columns  and 'months_as_customer' in df.columns:
        new_features['claim_per_month'] = df['total_claim_amount'] / (df['months_as_customer'] + 1)

    # found certain occupations more fraudulent
    risky_occupations = ['student', 'unemployed', 'self-employed']
    if 'insured_occupation' in df.columns:
        new_features['risky_occupation_flag'] = df['insured_occupation'].isin(risky_occupations).astype(int)

        #adding new features for train
    df=pd.concat([df,new_features],axis=1)

    return new_features
```
